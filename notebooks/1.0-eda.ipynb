```python
# ...existing code...
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load main data
df = pd.read_csv('../data/raw/data (1).csv')
# Load variable definitions
var_defs = pd.read_csv('../data/raw/Xente_variable_Definitions (1).csv')
```

```python
# Preview data
print(df.head())
print(df.info())
# Preview variable definitions
print(var_defs.head())
```

```python
# 2. Summary Statistics
print(df.describe())
```

```python
# 3. Numerical Feature Distribution
num_cols = df.select_dtypes(include='number').columns
df[num_cols].hist(figsize=(12, 8))
plt.show()
```

```python
# 4. Categorical Feature Distribution
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print(df[col].value_counts())
```

```python
# 5. Correlation Analysis
sns.heatmap(df[num_cols].corr(), annot=True)
plt.show()
```

```python
# 6. Missing Values
print(df.isnull().sum())
```

```python
# 7. Outlier Detection
for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```
